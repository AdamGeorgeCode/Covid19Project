```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots created
```

```python
filename = "Data - movies.csv"
```

```python
df = pd.read_csv('.\\' + filename)
```

```python
df.head()
```

```python
# See if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for the columns

df.dtypes
```

```python
# Drop null values

df = df.dropna(axis = 0, how = 'any')
```

```python
# Change data types of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# Create the correct year column - as deviation between current 'year' value and year in 'released'

yc = df['released'].astype(str)

yc = yc.str.split(' ').str.get(2)

yc = yc.str.strip(' ')

df['yearcorrect'] = yc
```

```python
df
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Correlation Hypothesis

# Budget high correlation with gross
# Company high correlation with gross
```

```python
# Scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot of budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```

```python
# Looking at correlation values
```

```python
df.corr(method = 'pearson')
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Numbers')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at company correlation

df.head()
```

```python
df_numerized = df
```

```python
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
```

```python
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Numbers')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr(method = 'pearson')
```

```python
correlation_mat = df_numerized.corr(method = 'pearson')

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
spdf = pd.DataFrame(sorted_pairs, columns = ['Correlation'])
```

```python
spdf
```

```python
high_corr = spdf[(spdf['Correlation'] > 0.5) & (spdf['Correlation'] < 1)]

high_corr
```
